```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
confirmed_global=pd.read_csv('time_series_covid19_confirmed_global.csv')
deaths_global=pd.read_csv('time_series_covid19_deaths_global.csv')
recovered_global=pd.read_csv('time_series_covid19_recovered_global.csv')

print('shape of confirmed_global', confirmed_global.shape)
print('shape of global death rate', deaths_global.shape)
print('shape of global recovery', recovered_global.shape)
```

```python
print(confirmed_global.columns)
print(deaths_global.columns)
print(recovered_global.columns)
```

```python
data_confirmed=pd.melt(confirmed_global, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
                      var_name=['Date'])
data_deaths_global=pd.melt(deaths_global, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
                          var_name=['Date'])
data_recovery_global=pd.melt(recovered_global, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
                            var_name=['Date'])
```

```python
data_confirmed['Date']=pd.to_datetime(data_confirmed['Date'])
data_deaths_global['Date']=pd.to_datetime(data_deaths_global['Date'])
data_recovery_global['Date']=pd.to_datetime(data_recovery_global['Date'])

data_confirmed.head()
```

```python
data_confirmed.columns=data_confirmed.columns.str.replace('value','Confirmed')
data_deaths_global.columns=data_deaths_global.columns.str.replace('value','Deaths')
data_recovery_global.columns=data_recovery_global.columns.str.replace('value','Recovery')

data_confirmed.head()
```

```python
print(data_confirmed.columns)
print(data_deaths_global.columns)
print(data_recovery_global.columns)
```

```python
## check missing values
print(data_confirmed['Province/State'].unique())
df=data_confirmed['Province/State'].unique()
dd=len(df)
print(dd)
data_confirmed.isnull().sum()
```

```python
## handle missing values
data_confirmed['Province/State'].fillna(data_confirmed['Country/Region'], inplace=True)
data_deaths_global['Province/State'].fillna(data_deaths_global['Country/Region'], inplace=True)
data_recovery_global['Province/State'].fillna(data_recovery_global['Country/Region'], inplace=True)

data_confirmed.head()
```

```python
data_confirmed['Lat'].fillna(data_confirmed['Lat'].mean(), inplace=True)
data_confirmed['Long'].fillna(data_confirmed['Long'].mean(), inplace=True)
data_confirmed.isnull().sum()
```

```python
data_deaths_global['Lat'].fillna(data_deaths_global['Lat'].mean(), inplace=True)
data_deaths_global['Long'].fillna(data_deaths_global['Long'].mean(), inplace=True)
data_deaths_global.isnull().sum()
```

```python
data_recovery_global['Lat'].fillna(data_recovery_global['Lat'].mean(), inplace=True)
data_recovery_global['Long'].fillna(data_recovery_global['Long'].mean(), inplace=True)
data_recovery_global.isnull().sum()
```

```python
##Join data

full_join=data_confirmed.merge(data_deaths_global[['Province/State', 'Country/Region', 'Date', 'Deaths']],
                              how='left',
                              left_on=['Province/State', 'Country/Region', 'Date'],
                              right_on=['Province/State', 'Country/Region', 'Date'])
full_join.head()
```

```python
full_join=full_join.merge(data_recovery_global[['Province/State', 'Country/Region', 'Date', 'Recovery']],
                         how='left',
                         left_on=['Province/State', 'Country/Region', 'Date'],
                         right_on=['Province/State', 'Country/Region', 'Date'])
full_join.head()
```

```python
full_join['month-year']=full_join['Date'].dt.strftime('%b-%y')
full_join.head()
```

```python
full_join2=full_join.copy()

full_join2['Date - 1']=full_join2['Date']+pd.Timedelta(days=1)
full_join2.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovery': 'Recovered - 1',
                          'Date': 'Date Minus 1'}, inplace=True)

full_join2
```

```python
full_join3 = full_join.merge(full_join2[['Province/State', 'Country/Region','Confirmed - 1', 'Deaths - 1', 
                            'Recovered - 1', 'Date - 1', 'Date Minus 1']], how = 'left',
                             left_on = ['Province/State','Country/Region','Date'], 
                             right_on = ['Province/State', 'Country/Region','Date - 1'])
full_join3.head()
```

```python
full_join3['Confirmed Daily'] = full_join3['Confirmed'] - full_join3['Confirmed - 1']
full_join3['Deaths Daily'] = full_join3['Deaths'] - full_join3['Deaths - 1']
full_join3['Recovered Daily'] = full_join3['Recovery'] - full_join3['Recovered - 1']

print(full_join3.shape)
```

```python
full_join3.head()
```

```python
#creating a new df    
full_join2 = full_join.copy()

#creating a new date columns - 1
full_join2['Date - 1'] = full_join2['Date'] + pd.Timedelta(days=1)
full_join2.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovery': 'Recovered - 1',
                          'Date': 'Date Minus 1'}, inplace=True)
full_join2.head()
```

```python
#Joing on the 2 DFs
full_join3 = full_join.merge(full_join2[['Province/State', 'Country/Region','Confirmed - 1', 'Deaths - 1', 
                            'Recovered - 1', 'Date - 1', 'Date Minus 1']], how = 'left',
                             left_on = ['Province/State','Country/Region','Date'], 
                             right_on = ['Province/State', 'Country/Region','Date - 1'])

#minus_onedf.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovered': 'Recovered - 1'}, inplace=True)
full_join3.head()

# Additional Calculations
full_join3['Confirmed Daily'] = full_join3['Confirmed'] - full_join3['Confirmed - 1']
full_join3['Deaths Daily'] = full_join3['Deaths'] - full_join3['Deaths - 1']
full_join3['Recovered Daily'] = full_join3['Recovery'] - full_join3['Recovered - 1']

print(full_join3.shape)
```

```python
full_join3.head()
```

```python
full_join3['Confirmed Daily'].loc[full_join3['Date']== '2020-01-22']= full_join3['Confirmed']
full_join3['Deaths Daily'].loc[full_join3['Date']== '2020-01-22']= full_join3['Deaths']
full_join3['Recovered Daily'].loc[full_join3['Date']== '2020-01-22']= full_join3['Recovery']

del full_join3['Confirmed - 1']
del full_join3['Deaths - 1']
del full_join3['Recovered - 1']
del full_join3['Date - 1']
del full_join3['Date Minus 1']
```
